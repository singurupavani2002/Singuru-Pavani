**Load Dataset**

```python
df = pd.read_excel("CLEANED DATASET.xlsx")
df.head()
```

**BAR CHART**

```python
df['type'].value_counts().plot(kind='bar')
plt.title("Movies vs TV Shows")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()
```

**LINE CHART**

```python
year_count = df['release_year'].value_counts().sort_index()

plt.plot(year_count.index, year_count.values)
plt.title("Content Released Over Years")
plt.xlabel("Year")
plt.ylabel("Number of Shows")
plt.show()
```

**PIE CHART**

```python
df['type'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Content Distribution")
plt.ylabel("")
plt.show()
```

**HISTOGRAM**

```python
plt.hist(df['release_year'].dropna())
plt.title("Release Year Distribution")
plt.xlabel("Year")
plt.ylabel("Frequency")
plt.show()
```

**SCATTER PLOT**

```python
# Convert duration to numeric (if needed)
df['duration_num'] = df['duration'].str.extract('(\d+)').astype(float)

plt.scatter(df['release_year'], df['duration_num'])
plt.title("Year vs Duration")
plt.xlabel("Release Year")
plt.ylabel("Duration")
plt.show()
```

**BOX PLOT**

```python
plt.boxplot(df['duration_num'].dropna())
plt.title("Duration Box Plot")
plt.ylabel("Duration")
plt.show()
```

**AREA CHART**

```python
year_count.sort_index().plot(kind='area')
plt.title("Year-wise Growth")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```
