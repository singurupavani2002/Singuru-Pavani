**2.Handle Missing Values**

```python
df_cleaned = df.copy()


df_cleaned['date_added'] = pd.to_datetime(df_cleaned['date_added'], errors='coerce')


for col in df_cleaned.columns:
    if pd.api.types.is_object_dtype(df_cleaned[col]):
        df_cleaned[col] = df_cleaned[col].fillna('')
    elif pd.api.types.is_numeric_dtype(df_cleaned[col]):
        df_cleaned[col] = df_cleaned[col].fillna(0)
    elif pd.api.types.is_datetime64_any_dtype(df_cleaned[col]):

        df_cleaned[col] = df_cleaned[col].astype(str).replace('NaT', '').fillna('')

print("Missing values after cleaning:")
display(df_cleaned.isnull().sum())
```

**Save Cleaned Data**

```python
df_cleaned.to_excel("numpy_cleaned.xlsx", index=False)
```

**3.DATA PREPARATION**

**Select Numeric Data**

```python
numeric_data = df_cleaned.select_dtypes(include=[np.number]).to_numpy()
```

**Normalize Data**

```python
norm_data = (numeric_data - np.min(numeric_data, axis=0)) / (np.max(numeric_data, axis=0) - np.min(numeric_data, axis=0))
```

**Standardization**

```python
mean = np.mean(numeric_data, axis=0)
```

```python
std = np.std(numeric_data, axis=0)
```

```python
standardized_data = (numeric_data - mean) / std
```

**Sort Data**

```python
sorted_data = standardized_data[standardized_data[:, 0].argsort()]
```

**Save Prepared Data**

```python
df_prepared = pd.DataFrame(sorted_data)
```

```python
df_prepared.to_excel("numpy_prepared.xlsx", index=False)
```
