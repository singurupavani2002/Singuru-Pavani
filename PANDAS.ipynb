**2. DATA CLEANING**

**Import Library**

```python
import pandas as pd
```

**Upload Dataset**

```python
from google.colab import files
uploaded = files.upload()
```

**Load Dataset**

```python
df = pd.read_excel("CLEANED DATASET.xlsx")
```

**View Data**

```python
df.head()
```

**Check Missing Values**

```python
df.isnull().sum()
```

**Handle Missing Values**

```python
df = df.dropna()
```

```python
df = df.fillna("Unknown")
```

**Remove Duplicates**

```python
df = df.drop_duplicates()
```

**Rename Columns**

```python
df.columns = df.columns.str.strip().str.lower()
```

**Check Clean Data**

```python
df.info()
```

**Save Cleaned Data**

```python
df.to_excel("cleaned_data.xlsx", index=False)
```

**3. DATA PREPARATION**

**Convert Date Column**

```python
df['date_added'] = pd.to_datetime(df['date_added'])
```

**Extract Year & Month**

```python
df['year_added'] = df['date_added'].dt.year
```

```python
df['month_added'] = df['date_added'].dt.month
```

**Create New Columns**

```python
df['duration_category'] = df['duration'].apply(lambda x: 'Short' if '90' in str(x) else 'Long')
```

**Group Data**

```python
df_grouped = df.groupby('genre')['show_id'].count()
```

**Sort Data**

```python
df = df.sort_values(by='release_year', ascending=False)
```

**Reset Index**

```python
df = df.reset_index(drop=True)
```

**Save Prepared Data**

```python
df.to_excel("prepared_data.xlsx", index=False)
```
